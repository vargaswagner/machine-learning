```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creamos una dataset aleatorio
data = np.random.uniform(size=(30,2))
print(data)
```

```python
plt.scatter(data[:, 0], data[:, 1])
```

```python
from sklearn.cluster import KMeans
```

```python
# Creamos 4 centroides con k=4

numero_clusters = int(input('Ingrese el numero de grupos a CREAR:'))
kmedias = KMeans(n_clusters=numero_clusters, n_init=10).fit(data)
centroides = kmedias.cluster_centers_
etiquetas = kmedias.labels_
# print(centroides)
# print(etiquetas)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=kmedias.labels_)
plt.scatter(centroides[:, 0], centroides[:, 1], s=100, c='red')
```

```python
# Usamos la m√©trica de suma de errores para diferentes valores de K (de grupos)

rendimiento = []

for r in range(1,10):
  kmedias = KMeans(n_clusters=r, n_init=10).fit(data)
  rendimiento.append(kmedias.inertia_)

rendimiento
```

```python
plt.plot(range(1,10), rendimiento, marker='o')
```
